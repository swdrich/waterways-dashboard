```python
# Import dependencies

import pandas as pd
```

```python
# Set up file paths for csv files

trip_path = "../Static/data/Cedar Lake Van Buren County Two - Sheet1.csv"

trip_data = pd.read_csv(trip_path)
trip_data.head()
```

```python
trip_data.dtypes
```

```python
# Rename columns
trip_data = trip_data.rename(columns={"Lake Name": "lake_name", "County": "county", "State": "state", "SurveyName": "survey_name", 
                          "Contributor Name": "contributor_name", "Year": "year", "Month": "month", "Day": "day",
                          "Date": "date", "Approx Time of Day": "timestamp", "Elapsed Time (sec)": "elapsed_time_seconds",
                          "Surface Water Temperature (oF)": "surface_water_temp_F", "Conductivity (µS/cm)": "conductivity", 
                          "Latitude": "latitude", "Longitude": "longitude", "Location Name": "location_name", "Notes": "notes"})
trip_data.head()
```

```python
# Combine date and time into timestamp column
trip_data.timestamp = trip_data.date + " " + trip_data.timestamp
trip_data.timestamp = pd.to_datetime(trip_data.timestamp)
trip_data.head()
```

```python
# Convert date to datetime
trip_data.date = pd.to_datetime(trip_data.date)
trip_data
```
